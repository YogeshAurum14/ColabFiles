```python

import re
v = ""
account_list = []
noneType = 0
date_list = []
time_list = []
for i in range(len(q)):
    v = q[i]
    a = re.search("X\w+\s",v)
    date_extrctr = re.search("\d+/\d+/\d+",v)
    time_extrctr = re.search("\d+:\d+:\d+",v)
    w = str(type(a))
    date_type = str(type(date_extrctr))
    time_type = str(type(time_extrctr))
    if((w == "<class 're.Match'>")  ):
#         print(a.group())
        account_list.append(a.group())
    elif(w =="<class 'NoneType'>"):
        noneType += 1
        continue
    elif((date_type=="<class 're.Match'>")):
        date_list.append(date_extrctr.group())
    elif(date_type=="<class 'NoneType'>"):
        continue
    elif((time_type=="<class 're.Match'>")):
        time_list.append(time_extrctr.group())
    elif(time_type=="<class 'NoneType'>"):
        continue
    else:
        print("unknown match type")

print(account_list)
print(noneType)
print(date_list)
print(time_list)
```

```python
def set_custom_boundaries1(doc):  
    for token in doc[:-1]:
        #adding a new segmentation rule
        if token.text == ")":
            #we traverse through the token and once we find ;, we indicate that from here new sentence starts
            doc[token.i+1].is_sent_start = True
    return doc
```

```python
nlp.add_pipe(set_custom_boundaries1, before = 'parser')
nlp.pipe_names
```
